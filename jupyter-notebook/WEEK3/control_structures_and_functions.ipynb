# if else family of constructs
- if, if-else, if-elif-else are a family of constructs where:
  - A condition is first checked, if it is satisfied then statements inside the body of construct are executed.
  - if condition is not satisfied, code exits construct or moves on to other options.

# For loop

# Funtions in Python:
- A function accepts input arguments and produces an output by executing valid commands present in the function body.
- Functions are created using the keyword def followed by the function name and a colon with the statements to be executed indented as a block.
  def function_name(parameters):
      statements

```python
# Converting the Age variable from months to years by defining a function.
# The converted values will be stored in a new column, 'Age_Converted'.
# inserting a new column
cars_data1.insert(11,'Age_Converted',0);
```

```python
def age_convert(age):
    return (age/12)
cars_data1['Age_Converted'] = age_convert(cars_data1['Age']);
cars_data1['Age_Converted'] = round(cars_data1['Age_Converted'],1)
```

```python
cars_data1.head()
```

```python
# Converting the Age variable from months to years and getting kilometers(KM) run per month.
# The converted values of kilometer will be stored in a new column, 'km_per_month'.
# inserting a new column
cars_data1.insert(12,'km_per_month',0)
```

```python
def convert(age,km):
    age_convert = age/12
    km_convert = km/age
    return [age_convert,km_convert]

cars_data1['Age_Converted'],cars_data1['km_per_month'] = convert(cars_data1['Age'],cars_data1['KM'])
```

```python
cars_data1.head()
```

```python
cars_data1['Age_Converted'] = round(cars_data1['Age_Converted'],1)
cars_data1['km_per_month'] = round(cars_data1['km_per_month'],1)
cars_data1.head()
```
